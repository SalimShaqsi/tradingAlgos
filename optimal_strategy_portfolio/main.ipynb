```python
from shared_utils.data import get_securities_data
from optimal_strategy_portfolio.signals.aggregators import HMean
from optimal_strategy_portfolio.signals.transforms import SMA
from optimal_strategy_portfolio.signals.price_signals import R
from optimal_strategy_portfolio.signals.variables import BoolVar, Var
from optimal_strategy_portfolio.signals.aggregators import HStd
from optimal_strategy_portfolio.solver_wrappers import pygad_wrapper
from optimal_strategy_portfolio.metrics import sharpe
import quantstats as qs
qs.extend_pandas()
```

```python
symbs = ['GOOG', 'AMZN', 'FB', 'TSLA', 'SPY']
data = get_securities_data(symbs)
```

```python
from optimal_strategy_portfolio.signals.price_signals import PriceMASignal

return_signals = [R(data[symb], name=f'{symb}_returns') for symb in symbs]
slow_sma_returns = [SMA(r) for r in return_signals]
fast_sma_returns = [SMA(r) for r in return_signals]
z = HMean(slow_sma_returns) + Var(bounds=((0, 5),)) * HStd(slow_sma_returns)

relative_crossover_buys = [(fast > z) for fast in fast_sma_returns]
relative_crossover_sells = [(fast < z) for fast in fast_sma_returns]

abs_crossover_buys = [(fast > slow) for fast, slow in zip(fast_sma_returns, slow_sma_returns)]
abs_crossover_sells = [(fast < slow) for fast, slow in zip(fast_sma_returns, slow_sma_returns)]

fast_sma_prices = [PriceMASignal(data[symb], name=f'{symb}_price_fast_sma') for symb in symbs]
slow_sma_prices = [PriceMASignal(data[symb], name=f'{symb}_price__slow_sma') for symb in symbs]

price_crossover_buys = [fast > slow for fast, slow in zip(fast_sma_prices, slow_sma_prices)]
price_crossover_sells = [fast < slow for fast, slow in zip(fast_sma_prices, slow_sma_prices)]

buys = [(BoolVar() | b1) * (BoolVar() | b2) * (BoolVar() | b3)
        for b1, b2, b3 in zip(relative_crossover_buys, abs_crossover_buys, price_crossover_buys)]
sells = [(BoolVar() | b1) * (BoolVar() | b2) * (BoolVar() | b3)
         for b1, b2, b3 in zip(relative_crossover_sells, abs_crossover_sells, price_crossover_sells)]

signal = buys[0] - sells[0]
```

```python
strat = signal.to_strategy()

r = strat.optimize(solver=pygad_wrapper, num_generations=500)

x = r[0]

returns_data = strat.execute_to_pandas(*x)
```

```python
returns_data.sharpe()

stock = qs.utils.download_returns('FB')

sharpe(returns_data.to_numpy())
```
